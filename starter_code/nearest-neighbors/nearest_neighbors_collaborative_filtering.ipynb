```python
# imports
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# sample code to read in pivot table 
# yay stackoverflow!
# https://stackoverflow.com/questions/55359807/save-pandas-pivot-table-to-include-index-and-columns-names

utility_matrix = pd.read_csv('../../data/user_rating_matrix.csv', index_col=0)
utility_matrix.columns.name = utility_matrix.index.name
utility_matrix.index.name = 'title'
utility_matrix.head()
```

```python
from sklearn.neighbors import NearestNeighbors

# knn = NearestNeighbors(algorithm='brute', metric='cosine')
knn = NearestNeighbors()
knn.fit(utility_matrix.values)

print(utility_matrix.values)
# feature_test = to_test['50_samples'].values
```

```python
# random_movie = np.random.choice(utility_matrix.shape[0])
random_movie = 4442

# Forrest Gump: 3158
# Iron Man: 4442
# Avatar: 688
# 
print("Movie is: ", utility_matrix.index[random_movie])

# print(utility_matrix.shape)
```

```python
feature_input = utility_matrix.iloc[random_movie: random_movie+2,:].values.reshape(2,-1) # + 2 for iron man 1 and 2
print(feature_input.shape)
# print(feature_input)

# distances, indices = knn.kneighbors(utility_matrix.iloc[random_movie].values.reshape(1,-1), n_neighbors = 11)
distances, indices = knn.kneighbors(feature_input, n_neighbors = 11)

print(indices)
print(distances)
```

```python
m=pd.Series(indices[0], name='movie')
d=pd.Series(distances[0], name='distance')

recommendations = pd.concat([m,d], axis=1).sort_values(by='distance', ascending=True)

print(recommendations)
print(utility_matrix.index[random_movie], '...makes sense that it is the closest movie since it is the one we used for query')
```
